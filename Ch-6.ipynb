```python
# if same character pass else display diff. character line & position.
with open("task2.txt") as f1 :
    s1=f1.read()
with open("task3.txt") as f2:
    s2=f2.read()

c=n=1
for i in range(len(s1)):
    if s1[i]==s2[i]:
        c+=1
        if s1[i]=="\n":
            n+=1
            c=1
    else:
        print("Line :",n,", position :",c)
        c+=1
f1.close()
f2.close()
```

```python
# wap to remove comment given file and write all line without comment new file.
f1 = open("task2.txt","r")
f2 = open("task3.txt","w")

while True:
    l=f1.readline()
    if len(l)!=0:
        if l[0]=="#":
            continue
        else:
            if "#" in l:
                for index in range(1,len(l)):
                    if l[index]=="#":
                        f2.write(l[0:index]+"\n")
            else:
                f2.write(l)
    else:
        break
f1.close()
f2.close()
```

```python
# wap to count the fqn of words from the given text file
f=open("d1.txt","r")
f1=f.read().split()
d={}
for i in f1:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
for i in d:
    print(i,"-",d[i],end=", ")
f.close()
```

```python
# wap to accept string or line from user quit the user enter end , if the enter line user should be new line....
# ....in files save all the line in text file and display only those lines which begins with capital letter.
f=open("task4.txt","w+")
while True:
    s=str(input("enter your data(for quit enter end) :"))
    if s.lower()=="end":
        break
    else:
        f.write(s+"\n")
f.seek(0)
f1=f.readlines()
for line in f1:
    if line[0].isupper():
        print(line[:-1])
f.close()
```

```python
# wap to read textfile and print only words starting with i in reverse order.
f=open("task4.txt","r")
data=f.read().split()
new=""
for word in data:
    if word[0]=="i" or word[0]=="I":
        new+=word[::-1]+" "
    else:
        new+=word+" "
print("your output :",new)
f.close()
```
