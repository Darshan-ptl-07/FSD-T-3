```python
class student:
    branch = 'CSE'
    # constuctor
    def __init__(self,rollno,name):
        self.rollno = rollno
        self.name = name
        
    # method
    def enter_marks(self):
        self.marks = []
        for i in range(5):
            self.marks.append(int(input(f"enter Marks of subject {i+1}: ")))
            
    # method
    def display(self):
        print("Name:",self.name)
        print("Branch:",self.branch)
        print("Roll No.:",self.rollno)
        if len(self.marks)>0:
            print("Marks:",self.marks)
            
    # method
    def calculate_avg(self):
        c=0
        sum=0
        for i in self.marks:
            sum+=i
            c+=1
        print("Avg. Marks:",sum/c)
```

```python
s1 = student(1,"Aryan")
s2 = student(2,"Dhairy")
s3 = student(3,"Kishan")
```

## WAP to create a class string and make a method to count uppercase,vowels,consonant and space

```python
class String:
    def __init__(self):
        self.uppercase=0
        self.lowercase=0
        self.vowels=0
        self.consonant=0
        self.spaces=0
        self.s = input("Enter your String :")
        
    def count_uppercase(self):
        for i in self.s:
            if i.isupper():
                self.uppercase+=1
    def count_lowercase(self):
        for i in self.s:
            if i.islower():
                self.lowercase+=1
    def count_vowels(self):
        s1 = "aeiouAEIOU"
        for i in self.s:
            if i in s1:
                self.vowels+=1
    def count_consonant(self):
        for i in self.s:
            if (i.isalpha()) and (i not in "aeiouAEIOU"):
                self.consonant+=1
                
    def count_spaces(self):
        for i in self.s:
            if i.isspace():
                self.spaces+=1
                
    def display(self):
        self.count_uppercase()
        self.count_lowercase()
        self.count_vowels()
        self.count_consonant()
        self.count_spaces()
        print("Total Uppercase = ",self.uppercase)
        print("Total Lowercase = ",self.lowercase)
        print("Total Vowels = ",self.vowels)
        print("Total consonant = ",self.consonant)
        print("Total Spaces = ",self.spaces)
```

```python
s=String()
```

```python
s.display()
```
