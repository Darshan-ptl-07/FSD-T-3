```python
%%writefile calculator.py
def add(a,b):
    print("sum:",a+b)
def sub(a,b):
    print("sub:",a-b)
def mul(a,b):
    print("mul:",a*b)
def div(a,b):
    print("div:",a/b)
def power(a,b):
    print("power:",a**b)
def module(a,b):
    print("module:",a%b)
def floor(a,b):
    print("floor div:",a//b)
```

```python
%%writefile mymodule.py
def mystring(s):
    print("Length:",len(s))
    print("Uppercase:",s.upper())
    print("Lowercase:",s.lower())
    print("Title:",s.title())
    print("Capitalize:",s.capitalize())
    print("is upper:",s.isupper())
    print("is lower:",s.islower())
    print("is title:",s.istitle())
    print("is alpha:",s.isalpha())
    print("is digit:",s.isdigit())
    print("is alnum:",s.isalnum())
def mylist(l):
    print("length:",len(l))
    l.append(50)
    print("append:",l)
    l.extend([10,20,30])
    print("extend:",l)
    l.sort()
    print("sort:",l)
    l.reverse()
    print("reverse:",l)
```

```python
import mymodule
n=eval(input("Enter Data : "))
if type(n)==str:
    mymodule.mystring(n)
if type(n)==list:
    mymodule.mylist(n)
```
